```python
import pandas as pd
url = 'https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=9e565f9a-84dd-4e79-9097-d403cae1ea75&1imit=1000&sort=ImportDate%20desc&format=CSV'
df = pd.read_csv(url)
df. head()
```

### 留下需要的欄位 （'sitename， 'county', 'pollutant'， 'status'， 'aqi'）

```python
df = df[['sitename', 'county', 'pollutant', 'status', 'aqi']]
df
```

### 找出空氣不良的測站（aqi > 100）

```python
condition = df['aqi'] > 100
df[condition]
```

### 查詢空氣指標污染物有哪些（pollutant）

```python
df['pollutant'].unique()
```

### 請找出高雄市的測站資料

```python
condition = df['county'] == '高雄市'
df[condition]
```

### 計算各縣市平均AQI

```python
df.groupby('county')['aqi'].mean().sort_values(ascending=False)
```
